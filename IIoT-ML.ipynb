```python
from sklearn.linear_model import LogisticRegression

X = [[30], [40], [50], [70]]
y = [0,0, 1, 1]

model = LogisticRegression()
model.fit(X, y)

print(model.predict([[60]]))
```

```python
from sklearn.linear_model import LinearRegression

sq = [[500], [800], [1000], [1500], [2000]] 
price = [20, 35, 45, 70, 95]                     

model = LinearRegression()   

model.fit(sq, price)

predicted_price = model.predict([[1700]])

print(predicted_price[0], "lakhs")
```

```python
from sklearn.linear_model import LinearRegression


X = [[500], [800], [1000], [1500], [2000]]
y = [20, 35, 45, 70, 95]

model = LinearRegression()
model.fit(X, y)

print("Predicted price for 1200 sq ft:", model.predict([[1200]])[0])
```

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

sq = [[500], [800], [1000], [1500], [2000]]
price = [20, 35, 45, 70, 95]

model = LinearRegression()
model.fit(sq, price)


pred = model.predict([[1500]])
print("Predicted Price:", pred[0], "lakhs")


plt.scatter(sq, price) 
plt.plot(sq, model.predict(sq))
plt.scatter(1500, pred[0])    

plt.xlabel("Square Feet")
plt.ylabel("Price (Lakhs)")
plt.title("House Price Prediction")

plt.show()
```
